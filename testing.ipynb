```python

def delete_directory(directory):
    try:
        shutil.rmtree(directory)
    except OSError as e:
        print(f"Error deleting directory {directory}: {e}")
```

```python
#Delete all node{x} folders and database
num_nodes = 9
for i in range(1, num_nodes + 1):
    node_directory = f"Week6/node{i}"
    delete_directory(node_directory)

# Delete the database file (modify the path as needed)
database_file = "Week6/files.db"
if os.path.exists(database_file):
    os.remove(database_file)
    print("Deleted database files.db")
```

```python
import sqlite3

# Path to the SQLite database file
database_file = "Week6/files.db"

# Create a connection to the database (this will create the file if it doesn't exist)
connection = sqlite3.connect(database_file)
cursor = connection.cursor()

# Define the schema for the "file" table
create_table_sql = """
CREATE TABLE IF NOT EXISTS `file` (
    `id` INTEGER PRIMARY KEY AUTOINCREMENT,
    `filename` TEXT,
    `size` INTEGER,
    `content_type` TEXT,
    `storage_mode` TEXT,
    `storage_details` TEXT,
    `created` DATETIME DEFAULT CURRENT_TIMESTAMP
);
"""

# Execute the SQL statement to create the table
cursor.execute(create_table_sql)

# Commit the changes and close the connection
connection.commit()
connection.close()

print(f"SQLite database '{database_file}' created with 'file' table.")
```

```python
#Upload
import requests
import time
import csv
import sys

# API endpoint for file upload
upload_url = "http://localhost:9000/files_mp"

# List of file names to upload
file_names = ["Data/100KB.bin", "Data/1MB.bin", "Data/10MB.bin", "Data/100MB.bin"]

# Number of times to upload each file
num_uploads = 5

# List to store all upload times
all_upload_times = []

# Define the payload (form data) for each request
payload = {
    'storage': 'raid1'
}

# Create a CSV file to store upload times
upload_csv_file = "upload_times.csv"
with open(upload_csv_file, mode='w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['File', 'Upload No.', 'Upload Time (seconds)'])

    # Perform the uploads and record times
    for file_name in file_names:
        print(f"Uploading {file_name}:")
        file_upload_times = []
        for i in range(num_uploads):
            # Prepare the file to upload
            with open(file_name, 'rb') as file:
                files = {'file': (file_name, file.read())}

            # Record the start time
            start_time = time.time()

            # Make the POST request to upload the file with form data
            response = requests.post(upload_url, data=payload, files=files)

            # Record the end time
            end_time = time.time()

            # Calculate the upload time
            upload_time = end_time - start_time

            # Append the upload time to the list
            file_upload_times.append(upload_time)

            # Write upload time to CSV file
            csv_writer.writerow([file_name, i + 1, upload_time])

            # Print progress statement
            if (i + 1) % 5 == 0:
                print(f"Uploaded {i + 1}/{num_uploads} times for {file_name}")

        all_upload_times.append(file_upload_times)

print(f"Upload times saved to {upload_csv_file}")
```
