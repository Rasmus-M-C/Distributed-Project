```python
import json
import matplotlib.pyplot as plt
# Create a subplot with 3 columns and 1 row, sharing the y-axis
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15, 5), sharey=True)

file_names = ['Data/buddygroup_kill_results.json', 'Data/random_kill_results.json', 'Data/minset_kill_results.json'] 

for i, file_name in enumerate(file_names):
    N_values = [12, 24, 36]
    for j, N in enumerate(N_values):
       
        # Read and filter results from the JSON file for a specific N
        with open(file_name, 'r') as json_file:
            results = json.load(json_file)
            
            # Filter results based on the target N
            filtered_results = [result for result in results if result['N'] == N]

        # Extract data for plotting from the filtered results
        s_values = [result['s'] for result in filtered_results]
        percentage_values = [result['loss_percentage'] for result in filtered_results]

        # Plot the data on the corresponding subplot
        axs[j].plot(s_values, percentage_values, marker='o', linestyle='-', color=plt.cm.plasma(i/len(file_names)), label=file_name.split('_')[0].capitalize())
        axs[j].set_title(f'Percentage of lost files for N = {N}')
        axs[j].set_xlabel('s')
        axs[j].set_ylabel('Percentage')
        axs[j].grid(True)
        axs[j].legend()

# Adjust the spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```
