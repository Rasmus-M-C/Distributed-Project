```python
import matplotlib.pyplot as plt
import numpy as np
from math import comb

def F_r(S, N, k):
    return 1 - (1 - comb(S, k) / comb(N, k))**4

def F_min(S, N):
    return 1 - (1 - comb(S, 3) / comb(N, 3))

def F_buddy(S, N):
    return 1 - ((1 - comb(S, 3) / comb(N, 3))**3)

# Generate values for S
S_values = [2, 4, 6, 8, 10]

# Set values for N
N_values = [12, 24, 36]
k = 3

# Create subplots
fig, axs = plt.subplots(1, len(N_values), sharey=True, figsize=(15, 5))

# Iterate over N values and plot the functions
for i, N in enumerate(N_values):
    F_r_values = [F_r(S, N, k) for S in S_values]
    F_min_values = [F_min(S, N) for S in S_values]
    F_buddy_values = [F_buddy(S, N) for S in S_values]

    axs[i].plot(S_values, F_r_values, label=f'$F_r$, N={N}, k={k}',color=plt.cm.plasma(0.33))
    axs[i].plot(S_values, F_min_values, label=f'$F_{{min}}$, N={N}',color=plt.cm.plasma(0.66))
    axs[i].plot(S_values, F_buddy_values, label=f'$F_{{buddy}}$, N={N}',color=plt.cm.plasma(0))
    axs[i].set_xlabel('Number of failed nodes (S)')
    axs[i].grid(True)
    # Add legend to the last subplot
    axs[i].legend(loc='upper left')

# Set common ylabel and title
axs[0].set_ylabel('Probability')
fig.suptitle('Probability Functions for loss pr. Number of Failed Nodes')
plt.tight_layout()
# Display the plot
#plt.grid(True)
plt.show()
```
